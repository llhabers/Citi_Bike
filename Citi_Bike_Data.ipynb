```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as st
from scipy.stats import linregress
```

```python
# Load and Read CSVs
JC201510citibiketripdata = "Resources/JC-201510-citibike-tripdata.csv"
JC201701citibiketripdata = "Resources/JC-201701-citibike-tripdata.csv"

# Read the mouse data and the study results
citi_bike2015 = pd.read_csv(JC201510citibiketripdata)
citi_bike2017 = pd.read_csv(JC201701citibiketripdata)
```

```python
citi_bike2015.head()
```

```python
citi_bike2017.head()
```

```python
cb = [citi_bike2015, citi_bike2017]
```

```python
cb_new = pd.concat(cb)

cb_new
```

```python
city_bikes = cb_new.sort_values(by='Bike ID', ascending=True)
city_bikes
```

```python
cb = city_bikes.rename(columns={'Trip Duration':'Trip Duration (Sec)'})
cb
```

```python
cb = cb.fillna(0)
cb
```
